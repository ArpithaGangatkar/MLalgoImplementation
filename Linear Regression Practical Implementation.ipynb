```python
## 1. Divide  the dataset into independent and dependent
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=36) 
```

```python
X_train.shape,X_test.shape
```

```python
## 2. Feature Scaling - Standardisation
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
scaler.fit(X_train) # compute S.D by applying Z-score, no need fit test for X test, Scaler is an object
```

```python
X_train = scaler.transform(X_train)
```

```python
X_train
```

```python
X_test=scaler.transform(X_test)
```

```python
X_test
```

```python
## 3. Model training
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()
```

```python
## 4. Model Fit
regression.fit(X_train,y_train)
```

```python
## 5. Coeeficients and intercept
regression.coef_
```

```python
regression.intercept_
```

```python
## 6. Prediction
y_pred = regression.predict(X_test)
```

```python
y_pred
```

```python
## 7. MSE, MAE, RMSE
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np
```

```python
mse=mean_squared_error(y_test, y_pred)
print(mse)
mae=mean_absolute_error(y_test, y_pred)
print(mae)
print(np.sqrt(mse))
```

```python
## Accuracy r2 and adjusted r square
from sklearn.metrics import r2_score
```

```python
score = r2_score(y_test, y_pred)
```

```python
score
```

```python
# display adjusted R-squared
1- (1-score)*(len(y)-1)/(len(y)-X.shape[1]-1)
```

```python
from sklearn.linear_model import Ridge
ridge = Ridge(alpha=20.0)
ridge.fit(X_train, y_train)
```

```python
y_pred=ridge.predict(X_test)
```

```python
mse=mean_squared_error(y_test, y_pred)
print(mse)
mae=mean_absolute_error(y_test, y_pred)
print(mae)
print(np.sqrt(mse))
```

```python
from sklearn.linear_model import Lasso
```

```python
lasso = Lasso(alpha=20.0)
lasso.fit(X_train, y_train)
```

```python
y_pred=lasso.predict(X_test)
mse=mean_squared_error(y_test, y_pred)
print(mse)
mae=mean_absolute_error(y_test, y_pred)
print(mae)
print(np.sqrt(mse))
```

```python
from sklearn.linear_model import ElasticNet
```

```python
elasticnet=ElasticNet(alpha=20.0)
elasticnet.fit(X_train,y_train)
```

```python
y_pred=elasticnet.predict(X_test)
mse=mean_squared_error(y_test, y_pred)
print(mse)
mae=mean_absolute_error(y_test, y_pred)
print(mae)
print(np.sqrt(mse))
```
